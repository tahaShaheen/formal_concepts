```python
# This is the data sensei used in his notes to teach Formal Concepts and make Hasse Diagram

data = {'m_1': [1, 1, 1, 1],
        'm_2': [1, 1, 1, 1],
        'm_3': [1, 1, 1, 0],
        'm_4': [1, 1, 0, 1],
        'm_5': [1, 0, 1, 1],
        'm_6': [1, 0, 1, 1],
        'm_7': [1, 1, 0, 0],
        'm_8': [1, 1, 0, 0],
        'm_9': [1, 0, 0, 0],
        'm_10': [1, 0, 0, 0],
        'm_11': [0, 0, 1, 1],
        'm_12': [0, 0, 1, 1]
          }

df = pd.DataFrame(data=data)
df.index = ['g_1', 'g_2', 'g_3', 'g_4']

print(df)
```

```python
# Creating all the column combinations

column_combinations = list()
for k in range(len(df.columns) + 1):
    column_combinations.append(list(combinations(df.columns, k)))
```

```python
# Removing null combinations

column_combinations = [ elem for elem in column_combinations if len(elem[0]) > 0]
```

```python
# Keeping only combinations with m_1 in them
# m_1 is connected to all g members. Any formal concept that will come out will have it. 
# It is pointless to calculate for the cobinations that do not have m_1

# Removing this part of the code will not make a difference in the output 
# because subsets are removed eventually anyway

column_combinations_1 = list()
for combination in column_combinations:
    column_combinations_1.append([elem for elem in combination if elem[0] == 'm_1'])
    
column_combinations = column_combinations_1.copy()

len(column_combinations)
```

```python
# Creating the formal concepts by anding all the columns that are in the combination

formal_concepts = list()

for combinations in column_combinations:
    for combination in combinations:
        prev = df[column_combinations[0][0][0]]
        for elem in combination:
            prev = prev & df[elem]
        g = tuple(prev.loc[prev == 1].index.values)
        m = combination
        if (len(g)>0):
            formal_concepts.append([m, g])
        
formal_concepts
```

```python
# Must remove the formal concepts that are subsets of bigger formal concepts

formal_concepts_improved = formal_concepts.copy()

for formal_concept_i in formal_concepts:

#   So that it doesn't match with itself
    formal_concepts_missing_i = formal_concepts.copy()
    formal_concepts_missing_i.remove(formal_concept_i)
    

    for formal_concept_j in formal_concepts_missing_i:
        
        if(set(formal_concept_i[0]).issubset(formal_concept_j[0])):
            
            if((set(formal_concept_i[1]).issubset(formal_concept_j[1]))):
                print(formal_concept_i, formal_concept_j)
                formal_concepts_improved.remove(formal_concept_i)
                break
```
